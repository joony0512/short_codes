```python
import re

def extract_titles(file_path):
    titles = []
    
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        next_line = False  # 다음 줄을 읽어야 하는지 여부를 나타내는 플래그
        
        for i in range(len(lines)):
            line = lines[i].strip()
            
            # 다음 줄을 읽어야 할 경우
            if next_line:
                titles.append(line)  # 다음 줄을 제목으로 추가
                next_line = False  # 플래그 초기화
                continue
            
            # 패턴 매칭을 통한 제목 추출
            if re.search(r'회사개요', line):
                if i - 1 >= 0:
                    titles.append(lines[i - 1].strip())  # 이전 줄을 제목으로 추가
            
            # 다음 줄을 읽어야 할 경우
            # else :
            #     if re.search(r'팁스(TIPS) 창업기업 세부현황 D/B •', line) or re.search(r'• TECH INCUBATOR PROGRAM FOR STARTUP', line):
            #         if i + 1 < len(lines):
            #             next_line = True  # 다음 줄을 읽어야 함
            #             next_line_empty = not lines[i + 1].strip()  # 다음 줄이 비어 있는지 확인
                    
            #             if next_line_empty:
            #                 if i + 2 < len(lines):
            #                     titles.append(lines[i + 2].strip())  # 다음 다음 줄을 제목으로 추가
            #                     next_line = False  # 플래그 초기화
                
    return titles

# 텍스트 파일 경로
file_path = '/content/drive/MyDrive/노동연구원/teams_pdf/IR_KOR_ALL.txt'

# 제목 추출
titles_df = extract_titles(file_path)

# 추출된 제목 출력
for title in titles_df:
    print(title)
len(titles_df)
```

## 생년월일 재추출


```python
birth_empty =pd.read_excel('/content/drive/MyDrive/노동연구원/teams_pdf/생년월일빈회사.xlsx')
```

```python
birth_empty
```

```python
def extract_birth_year(file_path, company_name):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        for i, line in enumerate(lines):
            if company_name in line:
                birth_line = lines[i + 2]  # 회사 이름 다음 다음 줄의 생년 정보
                year = re.findall(r'\d{4}', birth_line)  # 생년 정보에서 네 자리 숫자 추출
                if year:
                    return year[0]
    return None  # 회사 이름을 찾지 못하거나 생년 정보가 없는 경우 None 반환
```

```python
file_path = '/content/drive/MyDrive/노동연구원/teams_pdf/IR_KOR_ALL.txt'
company_name = birth_empty['회사명']
birth=[]
for i in company_name:
  birth_year = extract_birth_year(file_path, i)
  birth.append(birth_year)
  # print(birth_year)
birth
```

```python
pd.DataFrame(birth).to_excel('/content/drive/MyDrive/노동연구원/teams_pdf/생년월일빈거채움.xlsx', index=False)
```

## 사업화 단계 빈거 재추출


```python
grade_empty =pd.read_excel('/content/drive/MyDrive/노동연구원/teams_pdf/사업화단계빈회사.xlsx')
```

```python
import pandas as pd

import re

def find_business_stage(text_file, company_name):
    with open(text_file, 'r') as file:
        lines = file.readlines()

    business_stage = None
    for i in range(len(lines)):
        line = lines[i].strip()
        if line == company_name:
            for j in range(i + 1, len(lines)):
                if re.search(r'사업화 단계', lines[j]):
                    business_stage = lines[j].strip()
                    if re.search(r'사업화 단계', business_stage):
                      business_stage = business_stage.replace('사업화 단계', '').strip()
                      break
                    
                  
            break

    return business_stage

# 사용 예시
# text_file = '텍스트파일.txt'
# company_name = '프라비바이오'

# 사용 예시
excel_file = '/content/drive/MyDrive/노동연구원/teams_pdf/사업화단계빈회사.xlsx'
text_file = '/content/drive/MyDrive/노동연구원/teams_pdf/IR_KOR_ALL.txt'
company_name = grade_empty['회사명']

grade_df =[]
for i in   company_name:
    business_stage = find_business_stage(text_file, i)
    if business_stage:
        # print(f"{i}: {business_stage}")
        grade_df.append(business_stage)
    else:
        grade_df.append(None)
```

```python
pd.DataFrame(grade_df).to_excel('/content/drive/MyDrive/노동연구원/teams_pdf/사업화단계빈거채움.xlsx', index=False)
```

## 후속투자 재추출


```python
def find_investment_info(text_file, company_name):
    with open(text_file, 'r') as file:
        lines = file.readlines()

    investment_info = None
    found_company = False
    found_performance = False

    for i, line in enumerate(lines):
        line = line.strip()
        if line == company_name:
            found_company = True
        elif found_company and line == '주요성과':
            found_performance = True
        elif found_performance and found_company and '후속투자' in line:
            next_line_index = i + 2
            if next_line_index < len(lines):
                investment_info = lines[next_line_index].strip()
                break

    return investment_info

# 텍스트 파일과 회사명을 입력으로 넣어서 후속투자 관련 정보를 찾습니다.
text_file = '/content/drive/MyDrive/노동연구원/teams_pdf/IR_KOR_ALL.txt'
company_name = '프라비바이오'  # 회사명
investment_info = find_investment_info(text_file, company_name)

print(investment_info)
```
