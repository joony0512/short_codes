# 금광
image.png

```python
T = int(input())

for t in range(T):
    n, m = map(int, input().split())
    gold = list(map(int, input().split()))
    gold_list = []

    for g in range(n):
        gold_list.append(gold[m*g:m*(g+1)])



    for j in range(1,m):
        for i in range(1,n):

            if i ==0 :
                left_up = 0
            else :
                left_up = gold_list[i-1][j-1]

            if i ==n-1 :
                left_down = 0
            else : left_down = gold_list[i+1][j-1]

            left = gold_list[i][j-1]
            gold_list[i][j] = gold_list[i][j] + max(left_up, left, left_down)
    result = 0
    for i in range(n):
        result = max(result, gold_list[i][m-1])
    print(result)
```

```python
# 가장 긴 부분 수열 문제로 치환
# m[j]값으로 수열이 끝날때 그 수열의 길이를 dp에 저장


n = int(input())

m = list(map(int, input().split()))
m.reverse()

dp = [1] * n

for i in range(1, n):
    for j in range(0, i):
        if m[j] < m[i]:
            dp[i] = max(dp[i], dp[j]+1)
print(n - max(dp))
```

# 타일 채우기
image.png

```python
n = int(input())

def dpp(n):
    dp = [0] * (n + 1)
    dp[0] = 1
    if n >= 2:
        dp[2] = 3

    if dp[n] != 0:
        return dp[n]

    for i in range(4, n + 1, 2):
        for j in range(1, (i // 2) + 1):
            if j == 1:
                dp[i] += dpp(i - j * 2) * 3
            else:
                dp[i] += dpp(i - j * 2) * 2

    return dp[n]
print(dpp(n))
```

# 타일 채우기 2


```python
n = int(input())

d= [[0,0] for _ in range (n+1)]
d[0] = [1,1]
d[1] = [1,2]
d[2] = [1,7]


def dp(n):

    if d[n][1] !=0:
        return d[n][1]
    else:
        d[n][1] = dp(n-1) *2 + dp(n-2)*3

        for i in range(3, n+1):
            d[i][0] =2
            d[n][1] += dp(n-i) * 2

        return d[n][1]

a, b = dp(n)
print((a*b)%1000000007)
```
