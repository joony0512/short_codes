```python
n, m = map(int, input().split())
```

```python
k = n
count = 0
while k >1:
    if k%m ==0:
        k = k//m

    else :
        k -= 1
    count +=1
print(count)


```

```python
s = list(map(int, input().split()))
```

```python
a = s[0]

for i in s[1:] :
    if s[0] ==0 or s[0]==1:
        a += i
    else:
        if i == 0 or i ==1 :
            a += i
        else :
            a *= i


print(a)
```
