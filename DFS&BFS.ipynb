# queue #선입선출

```python
from collections import deque

queue = deque()

queue.append(5)
queue.append(4)
queue.append(5)
queue.append(2)
queue.popleft()
queue.append(1)
queue.popleft()
queue.append(3)
```

```python
print(queue) #먼저 들어온것부터 출력
```

```python
queue.reverse()
```

```python
print(queue) # 나중에 들어온것부터 출력
```

## 재귀함수

```python
def recursive():
    print('재귀함수를 호출합니다.')
    recursive()
recursive()
# 무한루프로 최대깊이 넘으면 에러
```

```python
# 재귀함수 종료조건 설정
def recursive(i):
    if i ==100:
        return
    print(i,'번째 재귀함수에서',i+1, '번째 재귀함수를 호출합니다.')
    recursive(i+1)
    print(i,  '번째 재귀함수를 종료.')
recursive(1)
```

## 팩토리얼

```python
def factorial(n):
    if n<=1:
        return 1
    return n * factorial(n-1)
```

```python
factorial(5)
```

# DFS (Depth First Search)

image.png

image.png

```python
graph =[
    [], # index 0 사용 안함
    [2,3,8], # 1번노드와 연결된 노드
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]
visited = [False] * 9
```

```python
def dfs(graph, v, visited):
    visited[v] = True
    print(v, end='')
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i , visited)

```

```python
dfs(graph, 1, visited)
```

```python
graph =[
    [], # index 0 사용 안함
    [2,3,8], # 1번노드와 연결된 노드
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]
visited = [False] * 9
```

```python
from collections import deque

def bfs(graph, start, visited):
    queue = deque([start])

    visited[start] = True

    while queue:
        v = queue.popleft()
        print(v, end='')
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
```
